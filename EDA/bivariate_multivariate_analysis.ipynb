```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic= pd.read_csv("train.csv")
tips= sns.load_dataset('tips')
flights= sns.load_dataset("flights")
iris= sns.load_dataset("iris")
```

## Numerical-Numerical

### 1. Scatter plot

```python
tips.head()
```

```python
# bivariate 
sns.scatterplot(tips["total_bill"], tips["tip"], color="purple")
```

```python
# multivariate
plt.figure(figsize=(10,7))
sns.scatterplot(tips["total_bill"], tips["tip"], hue=tips["sex"],style=tips["smoker"], size=tips["size"],palette="mako")
```

### 2. Line Plot

```python
flights.head(3)
```

```python
flights_=flights.groupby("year")["passengers"].sum().reset_index()
flights_
```

```python
sns.lineplot(flights_["year"], flights_["passengers"], color="purple")
```

# Numerical- Categorical

### 1. Bar plot

```python
titanic.head(3)
```

```python
# bivariate
sns.barplot(titanic["Pclass"], titanic["Fare"],palette="BrBG")
```

```python
# multivariate
sns.barplot(titanic["Pclass"], titanic["Age"], hue= titanic["Sex"],palette="BrBG")
```

### 2. Box Plot

```python
# bivariate
sns.boxplot(titanic["Sex"], titanic["Age"], palette="BuPu")
```

```python
# multivariate
sns.boxplot(titanic["Sex"], titanic["Age"], hue= titanic["Survived"], palette="BuPu")
```

### 3. Dist plot

```python
sns.distplot(titanic["Age"], hist=False, color="green")
```

```python
# bivariate
sns.distplot(titanic[titanic["Survived"]==0]["Age"], hist=False, color="green")
sns.distplot(titanic[titanic["Survived"]==1]["Age"], hist=False, color="cyan")
```
