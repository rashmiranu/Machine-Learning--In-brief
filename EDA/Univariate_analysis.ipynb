```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data= pd.read_csv("train.csv")
data.head()
```

# Categorical Data

```python
categorical_col= data[["Survived", "Pclass", "Sex","SibSp","Embarked", "Parch"]]
```

### 1. count plot 

```python
i=1
fig, axes= plt.subplots(2,3, figsize=(17,10))
for column in categorical_col.columns:
    plt.subplot(2,3,i)
    sns.countplot(data[column], palette="BuGn_r")
#     plt.title(column)
    fig.tight_layout(pad=2.0)
    fig.subplots_adjust(top=0.97)
    i=i+1
    
plt.show()
```

### 2. Pie chart

```python
i=1
fig, axes= plt.subplots(2,3, figsize=(17,10))
for column in categorical_col.columns:
    plt.subplot(2,3,i)
    data[column].value_counts().plot(kind="pie", autopct="%.2f")
#     plt.title(column)
    fig.tight_layout(pad=2.0)
    fig.subplots_adjust(top=0.97)
    i=i+1
    
plt.show()
```

# Numerical Data

```python
numerical_cols= data[["Age", "Fare"]] 
```

### 1. Histogram

```python
i=1
fig, axes= plt.subplots(1,2, figsize=(12,4))
for column in numerical_cols.columns:
    plt.subplot(1,2,i)
    plt.hist(data[column], bins=10, color="Salmon")
    plt.title(column)
    fig.tight_layout(pad=2.0)
    fig.subplots_adjust(top=0.97)
    i=i+1
    
plt.show()
```

### 2. Distplot

```python
i=1
fig, axes= plt.subplots(1,2, figsize=(12,4))
for column in numerical_cols.columns:
    plt.subplot(1,2,i)
    sns.distplot(data[column], bins=10, color="LightSalmon")
    plt.title(column)
    fig.tight_layout(pad=2.0)
    fig.subplots_adjust(top=0.97)
    i=i+1
    
plt.show()
```

### 3. Boxplot

```python
i=1
fig, axes= plt.subplots(1,2, figsize=(12,4))
for column in numerical_cols.columns:
    plt.subplot(1,2,i)
    sns.boxplot(data[column], color="#9FE2BF")
    plt.title(column)
    fig.tight_layout(pad=2.0)
    fig.subplots_adjust(top=0.97)
    i=i+1
    
plt.show()
```
